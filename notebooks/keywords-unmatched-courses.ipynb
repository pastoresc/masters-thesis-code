```python
import pandas as pd
from collections import Counter
import re

# Load match result
df = pd.read_csv("course_category_matches.csv")

# Normalize 'Matched Categories' column
df["Matched Categories"] = df["Matched Categories"].fillna("").astype(str).str.strip()

# Filter unmatched courses
df_unmatched = df[df["Matched Categories"].isin(["None", "", "none", "NONE"])]

# Tokenize and count
tokens = []
for title in df_unmatched["Course Title"]:
    words = re.findall(r'\b\w+\b', title.lower())
    filtered = [word for word in words if len(word) > 2]
    tokens.extend(filtered)

# Count word frequencies
word_freq = Counter(tokens)
top_words = word_freq.most_common(30)

# Output table
df_top_words = pd.DataFrame(top_words, columns=["Word", "Frequency"])
display(df_top_words)

# Save to CSV
df_top_words.to_csv("keywords_unmatched_courses.csv", index=False)
```
